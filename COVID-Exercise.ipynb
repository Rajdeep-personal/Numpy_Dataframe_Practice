# 2. Write a Python program to get the latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise.

```python
### sorting the dataset by date present in the 'Last_Update' column in descending order so that latest data come upwards
### then grouping the information of confirmed, deaths, recovered and active cases by applying 'groupby' on 'Country_Region'
### column.
```

```python
a = covid.sort_values(by = 'Last_Update',ascending = 'False').groupby(by = 'Country_Region')[['Confirmed','Deaths','Recovered','Active']]
```

```python
a.first()
```

#### Another way

```python
covid.groupby(by='Country_Region')[['Confirmed','Deaths','Recovered','Active']].sum()
```

# 4. Write a Python program to get the Chinese province wise cases of confirmed, deaths and recovered cases of Novel Coronavirus (COVID-19).

```python
covid_china = covid[covid['Country_Region']== 'China']
covid_china.groupby(by=['Country_Region','Province_State'])[['Confirmed','Deaths','Recovered']].sum().reset_index()
```

# 5. Write a Python program to get the latest country wise deaths cases of Novel Coronavirus (COVID-19).

```python
covid.groupby(by = ['Country_Region'])[['Deaths']].sum().reset_index()
```

### Checking purpose for date ascending and descending order

```python
covid_canada = covid[covid['Country_Region']== 'Canada']
covid_canada.sort_values(by = 'Last_Update',ascending = 'False').groupby(by = ['Last_Update','Country_Region','Province_State'])[['Deaths']].sum().reset_index()
```
